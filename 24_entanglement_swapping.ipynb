# Swap (troca) de emaranhamento

Vamos considerar que Darwin prepara um par de qubits maximamente emaranhados:
\begin{equation}
|\Phi_{+}\rangle_{AC} = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle).
\end{equation}
Depois disso, Darwin envia o qubit $A$ para o laboratório da Alice e o qubit $C$ para o laboratório do Charlie. 
Daril também prepara um par de qubits em um estado maximamente emaranhado,
\begin{equation}
|\Phi_{+}\rangle_{C'B} = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle),
\end{equation}
e envia o qubit $C'$ para Charlie e o qubit $B$ para o laboratório do Bob.

Primeiramente, vamos analisar o estado total. Usando a relação entre a base computacional e a base de Bell,
\begin{equation}
\begin{cases} 
|\Phi_{+}\rangle = \frac{1}{\sqrt{2}}\big(|00\rangle+|11\rangle\big) \\ 
|\Phi_{-}\rangle = \frac{1}{\sqrt{2}}\big(|00\rangle-|11\rangle\big) \\ 
|\Psi_{+}\rangle = \frac{1}{\sqrt{2}}\big(|01\rangle+|10\rangle \big) \\ 
|\Psi_{-}\rangle = \frac{1}{\sqrt{2}}\big(|01\rangle-|10\rangle\big) 
\end{cases} \therefore
\begin{cases} 
|00\rangle = \frac{1}{\sqrt{2}}\big(|\Phi_{+}\rangle+|\Phi_{-}\rangle\big) \\ 
|11\rangle = \frac{1}{\sqrt{2}}\big(|\Phi_{+}\rangle-|\Phi_{-}\rangle\big) \\ 
|01\rangle = \frac{1}{\sqrt{2}}\big(|\Psi_{+}\rangle+|\Psi_{-}\rangle \big) \\ 
|10\rangle = \frac{1}{\sqrt{2}}\big(|\Psi_{+}\rangle-|\Psi_{-}\rangle\big) 
\end{cases},
\end{equation}

podemos escrever
\begin{align}
|\Psi\rangle_{ACC'B} & = |\Phi_{+}\rangle_{AC}|\Phi_{+}\rangle_{C'B} = \frac{1}{2}\big(|00\rangle + |11\rangle)(|00\rangle + |11\rangle)\big) \\
& = \frac{1}{2}\big(|0000\rangle_{ACC'B} + |0011\rangle_{ACC'B} + |1100\rangle_{ACC'B} + |1111\rangle)_{ACC'B}\big) \\
& = \frac{1}{2}\big(|00\rangle_{AB}|00\rangle_{CC'} + |01\rangle_{AB}|01\rangle_{CC'} + |10\rangle_{AB}|10\rangle_{CC'} + |11\rangle_{AB}|11\rangle_{CC'}\big) \\
& = \frac{1}{2\sqrt{2}}\big(|00\rangle_{AB}\big(|\Phi_{+}\rangle_{CC'}+|\Phi_{-}\rangle_{CC'}\big) + |01\rangle_{AB}\big(|\Psi_{+}\rangle_{CC'}+|\Psi_{-}\rangle_{CC'}\big) \\
& \hspace{1.5cm} + |10\rangle_{AB}\big(|\Psi_{+}\rangle_{CC'}-|\Psi_{-}\rangle_{CC'}\big) + |11\rangle_{AB}\big(|\Phi_{+}\rangle_{CC'}-|\Phi_{-}\rangle_{CC'}\big)\big) \\
& = \frac{1}{2\sqrt{2}}\big(|\Phi_{+}\rangle_{CC'}\big(|00\rangle_{AB}+|11\rangle_{AB}\big) + |\Psi_{+}\rangle_{CC'}\big(|01\rangle_{AB}+|01\rangle_{AB}\big) \\
& \hspace{1.5cm} + |\Psi_{-}\rangle_{CC'}\big(|01\rangle_{AB}-|10\rangle_{AB}\big) + |\Phi_{-}\rangle_{CC'}\big(|00\rangle_{AB}-|11\rangle_{AB}\big)\big) \\
& = \frac{1}{2}\big(|\Phi_{+}\rangle_{CC'}|\Phi_{+}\rangle_{AB} + |\Psi_{+}\rangle_{CC'}|\Psi_{+}\rangle_{AB} + |\Psi_{-}\rangle_{CC'}|\Psi_{-}\rangle_{AB} + |\Phi_{-}\rangle_{CC'}|\Phi_{-}\rangle_{AB}\big) \\
\end{align}

Vemos assim que se Charlie faz medidas na base computacional, ele obterá um dos estados da base de Bell, que são equiprováveis, $Pr(\Phi_{\pm})=Pr(\Psi_{\pm})=1/4$. Para cada uma das alternativas, o estado dos qubits de Alice e Bob colapsará para o estado de Bell correspondente:
\begin{equation}
|\Psi\rangle_{ACC'B} \xrightarrow[]{MBB} 
\begin{cases}
|\Phi_{+}\rangle_{CC'} \rightarrow |\Phi_{+}\rangle_{AB}, \\
|\Psi_{+}\rangle_{CC'} \rightarrow |\Psi_{+}\rangle_{AB}, \\
|\Psi_{-}\rangle_{CC'} \rightarrow |\Psi_{-}\rangle_{AB}, \\
|\Phi_{-}\rangle_{CC'} \rightarrow |\Phi_{-}\rangle_{AB}.
\end{cases}
\end{equation}
Vemos assim, que os qubits de Alice e Bob terminam maximamente emaranhados mesmo sem nunca interagir diretamente.

Seguindo, consideremos que Alice, Bob e Charlie querem que o estado final de $AB$ seja o estado singleto:
\begin{equation}
|\Psi_{-}\rangle_{AB} = 2^{-1/2}\big(|01\rangle-|10\rangle\big),
\end{equation}
que é um estado que tem propriedades interessantes de invariância por rotações locais, que o torna possivelmente mais robusto à decoerência em comparação ao estado tripleto. Notemos ainda que os estados do triplepo podem ser levados, a menos de uma fase global, ao estado singleto pelas seguintes operações unitárias locais (__exercício:__ verifique):
\begin{align}
(I\otimes X)|\Phi_{-}\rangle_{AB} = |\Psi_{-}\rangle_{AB}, \\
(I\otimes ZX)|\Phi_{+}\rangle_{AB} = |\Psi_{-}\rangle_{AB}, \\
(I\otimes Z)|\Psi_{+}\rangle_{AB} = -|\Psi_{-}\rangle_{AB}.
\end{align}

Usemos também as transformações unitárias que levam da base de Bell para a base padrão:
\begin{equation}
(H\otimes I)C_{x}^{01}|B_{jk}\rangle = |jk\rangle \text{, para } j,k=0,1, 
\end{equation}
e
\begin{equation}
|B_{00}\rangle = |\Phi_{+} \rangle \text{, } |B_{01}\rangle = |\Psi_{+}\rangle \text{, } 
|B_{10}\rangle = |\Phi_{-}\rangle \text{, } 
|B_{11}\rangle = |\Psi_{-}\rangle.
\end{equation}
Com isso vemos que o circuito quântico abaixo simula, de forma totalmente unitária (coerente) o protocolo de troca de emaranhamento.

```python
import qiskit
qc = qiskit.QuantumCircuit(4, 2)
qc.h([0, 2]); qc.cx([0], [1]); qc.cx([2], [3]) # preparacao e distribuicao de emaranhamento
qc.barrier() # base de Bell -> base computacional
qc.cx([1], [2]); qc.h([1]) 
qc.barrier() # feedback baseado na info classica
qc.x([1, 2]); qc.cx([2], [3]); qc.cz([1], [3]); qc.x([1, 2]) 
qc.barrier() # para verificacao da funcionalidade do cq (tem que sair Psi-=11)
qc.cx([0], [3]); qc.h([0]); qc.measure([0, 3], [0, 1])
qc.draw(output = 'mpl')
```

## Entanglement swapping with partially entangled states
Let's consider three labs, where are Alice, Bob, and Charlie, respectively. Two pure states $$|\xi\rangle_{AB}=\sqrt{p}|00\rangle_{AB}+\sqrt{1-p}|11\rangle_{AB}$$ and $$|\eta\rangle_{CD}=\sqrt{q}|00\rangle_{CD}+\sqrt{1-q}|11\rangle_{CD}$$ are prepared, by someone. The quantons B and C are sent to Charlie, who makes a selective Bell basis measurement (BBM) on them. The quanton A is sent to Alice and the quanton D is sent to Bob. A the end of the protocol, A and D, that have never met before, can get entangled. Let's do the math for the states:
\begin{align}
|\xi\rangle_{AB}\otimes|\eta\rangle_{CD} &= \big(\sqrt{p}|00\rangle_{AB}+\sqrt{1-p}|11\rangle_{AB}\big)\otimes\big(\sqrt{q}|00\rangle_{CD}+\sqrt{1-q}|11\rangle_{CD}\big) \\
& = \sqrt{pq}|0000\rangle_{ABCD} + \sqrt{p(1-q)}|0011\rangle_{ABCD} + \sqrt{(1-p)q}|1100\rangle_{ABCD} + \sqrt{(1-p)(1-q)}|1111\rangle_{ABCD} \\
& = \sqrt{pq}|0\rangle_{A}\frac{1}{\sqrt{2}}\big(|\Phi_{+}\rangle+|\Phi_{-}\rangle\big)_{BC}|0\rangle_{D} + \sqrt{p(1-q)}|0\rangle_{A}\frac{1}{\sqrt{2}}\big(|\Psi_{+}\rangle+|\Psi_{-}\rangle \big)_{BC}|1\rangle_{D} \\ 
& \hspace{0.5cm} + \sqrt{(1-p)q}|1\rangle_{A}\frac{1}{\sqrt{2}}\big(|\Psi_{+}\rangle-|\Psi_{-}\rangle\big)_{BC}|0\rangle_{D} + \sqrt{(1-p)(1-q)}|1\rangle_{A}\frac{1}{\sqrt{2}}\big(|\Phi_{+}\rangle-|\Phi_{-}\rangle\big)_{BC}|1\rangle_{D} \\
& = |\Phi_{+}\rangle_{BC}\frac{1}{\sqrt{2}}\big(\sqrt{pq}|00\rangle_{AD}+\sqrt{(1-p)(1-q)}|11\rangle_{AD}\big) +  |\Phi_{-}\rangle_{BC}\frac{1}{\sqrt{2}}\big(\sqrt{pq}|00\rangle_{AD}-\sqrt{(1-p)(1-q)}|11\rangle_{AD}\big) \\
& \hspace{0.5cm} + |\Psi_{+}\rangle_{BC}\frac{1}{\sqrt{2}}\big(\sqrt{p(1-q)}|01\rangle_{AD}+\sqrt{(1-p)q}|10\rangle_{AD}\big) + |\Psi_{-}\rangle_{BC}\frac{1}{\sqrt{2}}\big(\sqrt{p(1-q)}|01\rangle_{AD}-\sqrt{(1-p)q}|10\rangle_{AD}\big).
\end{align}

Post-measurement states:
\begin{equation}
\begin{cases} |\Phi_{\pm}\rangle_{BC} \Rightarrow |\phi_{\pm}\rangle_{AD} = \frac{1}{\sqrt{2}}\big(\sqrt{pq}|00\rangle_{AD}\pm\sqrt{(1-p)(1-q)}|11\rangle_{AD}\big) \therefore \langle\phi_{\pm}|\phi_{\pm}\rangle=\frac{1}{2}(pq+(1-p)(1-q)), \\
|\Psi_{\pm}\rangle_{BC} \Rightarrow |\psi_{\pm}\rangle_{AD} = \frac{1}{\sqrt{2}}\big(\sqrt{p(1-q)}|01\rangle_{AD}\pm\sqrt{(1-p)q}|10\rangle_{AD}\big) \therefore \langle\psi_{\pm}|\psi_{\pm}\rangle=\frac{1}{2}(p(1-q)+(1-p)q). \end{cases}
\end{equation}

Normalized post-measurement states:
\begin{equation}
\begin{cases} |\Phi_{\pm}\rangle_{BC} \ \Rightarrow \ |\hat{\phi}_{\pm}\rangle_{AD} = \frac{\big(\sqrt{pq}|00\rangle_{AD}\pm\sqrt{(1-p)(1-q)}|11\rangle_{AD}\big)}{\sqrt{pq+(1-p)(1-q)}}, \\
|\Psi_{\pm}\rangle_{BC} \ \Rightarrow \ |\hat{\psi}_{\pm}\rangle_{AD} = \frac{\big(\sqrt{p(1-q)}|01\rangle_{AD}\pm\sqrt{(1-p)q}|10\rangle_{AD}\big)}{\sqrt{p(1-q)+(1-p)q}}. 
\end{cases}
\end{equation}

It is easy to see that if we set $p=q=1/2$, the probability of any of the BBM outcomes is 1/4 and the quantons A and D end up in a Bell (maximally entangle) state. 

In the geral case, let's recall that for a bipartite pure state $|\Phi\rangle=\alpha_{00}|00\rangle+\alpha_{01}|01\rangle+\alpha_{10}|10\rangle+\alpha_{11}|11\rangle$ the entanglement concurrence can be written as 
$$E_{con}(|\Phi\rangle)=|\langle\Phi|\sigma_{y}\otimes\sigma_{y}|\Phi^{*}\rangle|=2|\alpha_{00}\alpha_{11}-\alpha_{01}\alpha_{10}|.$$
So, for the protocol above, we shall have
\begin{align}
|\Phi_{\pm}\rangle_{BC} & \Rightarrow E_{conc}(|\hat{\phi}_{\pm}\rangle_{AD}) = 2|\pm\sqrt{pq(1-p)(1-q)}/(pq+(1-p)(1-q))-0| \\
& = 2\sqrt{p(1-p)q(1-q)}\big/(pq+(1-p)(1-q)), \\ 
|\Psi_{\pm}\rangle_{BC} & \Rightarrow E_{conc}(|\hat{\psi}_{\pm}\rangle_{AD}) = 2|0\mp\sqrt{p(1-q)q(1-p)}/(p(1-q)+(1-p)q)| \\
& = 2\sqrt{p(1-p)q(1-q)}\big/(p(1-q)+(1-p)q).
\end{align}
Below are shown plots of the entanglement of AD.

```python
import math

def E_Phi(p, q):
    return 2*math.sqrt(p*(1-p)*q*(1-q))/(p*q+(1-p)*(1-q))
def E_Psi(p, q):
    return 2*math.sqrt(p*(1-p)*q*(1-q))/(p*(1-q)+(1-p)*q)
```
